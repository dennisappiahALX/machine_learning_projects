```python
# Converting Each column to appropriate datatypes
date_cols = ['GarageYrBlt', 'YrSold','YearBuilt', 'YearRemodAdd']
cat_cols = [col for col in df.columns if df.dtypes[col]== 'object']
num_cols = [col for col in df.columns if col not in date_cols + cat_cols + ['Id','SalePrice']]
```

```python
# Converting Each column to appropriate datatypes
for col in df.columns:
  if col in date_cols:
    df[col] = pd.to_datetime(df[col])
  elif col in cat_cols:
    df[col] = df[col].astype('object')
```

```python
## MIssing Values Treatments

# For cat cols and date cols fill in with mode 
for col in df.columns:
  if col in date_cols + cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
    
  elif col in num_cols:
    df [col] = df[col].fillna(df[col].fillna(9999))
      
# Confirm that there aren't any missing values
df[df.columns.difference(['SalePrice'])].isna().sum()
```

```python
X_train = train_df.drop(columns= ['SalePrice'])
y_train = train_df.SalePrice
```

```python
from sklearn.feature_selection import mutual_info_regression
```

```python
discrete_features = X_train.dtypes == int
```

```python
def make_mi_scores(X_train, y_train, discrete_features):
    mi_scores = mutual_info_regression(X_train, y_train, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X_train.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X_train, y_train, discrete_features)
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(20, 20))
plot_mi_scores(mi_scores)  
```

```python
from sklearn.feature_selection import SelectKBest

selector = SelectKBest(mutual_info_regression, k= 35)
selector.fit(X_train, y_train)
X_train.columns[selector.get_support()]
```

```python
#Hyperparameter tuning

params = {
    "n_estimators" : [100, 500, 700, 900],
    "base_score"     : [0.25, 0.5, 0.75, 1],
    "learning_rate"   : [0.05, 0.1, 0.15, 0.2],
    "booster"        : ['gbtree','gblinear'],
    "max_depth"         : [2,3, 5, 10],
    "min_child_weight"  : [1,3,5,7],
}
```

```python
regressor = xgb.XGBRegressor()

random_search = RandomizedSearchCV(estimator= regressor, param_distributions = params , cv=5, n_iter = 50, 
                                   scoring ='neg_mean_absolute_error', verbose = 5, return_train_score = True, random_state=42)

random_search.fit(X_train, y_train)
```

```python
regressor = xgb.XGBRegressor(base_score=1, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
             importance_type='gain', interaction_constraints='',
             learning_rate=0.05, max_delta_step=0, max_depth=3,
             min_child_weight=3,monotone_constraints='()',
             n_estimators=700, n_jobs=8, num_parallel_tree=1, random_state=0,
             reg_alpha=0, reg_lambda=1, scale_pos_weight=1, subsample=1,
             tree_method='exact', validate_parameters=1, verbosity=None)
```

```python
from sklearn.model_selection import cross_val_score
```
